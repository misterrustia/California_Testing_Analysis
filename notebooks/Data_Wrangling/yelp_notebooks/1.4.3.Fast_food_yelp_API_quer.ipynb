```python
## testimng ----
## function to call and return data from yelp api 

def park_search(term,zipc,radius,lat,long):
    num_g = 0
    dist_g=0
    limit =50
    
    if zipc >0:
        location = int(zipc)
    else:
        location = zipc
        

    # no zip means all null ( district averages) 
    if lat  >0:
#         print('there is lat ',lat,' ',long)
        try:
            response = yelp_api.search_query(term = term,
                                       latitude=lat,
                                     longitude=long,
                                     radius = radius,
                                     limit = search_limit)
        except:
#             print('call failed')
            num_g = np.nan
            dist_g=np.nan
            return num_g , dist_g
            
        num_g = len(response['businesses'])
        if num_g>0:
            dist_g = np.round(response['businesses'][0]['distance'],0)
        else:
            dist_g = np.nan
        return num_g,dist_g
      
    
    elif location >0: # use zipcode 
#         print('use zip code',location)
        
        try:
            response = yelp_api.search_query(term = term,
                                             location=zipc,
                                     radius = radius,
                                     limit = search_limit)
        except:
#             print('call failed')
            num_g = np.nan
            dist_g=np.nan
            return num_g , dist_g
            

        num_g = len(response['businesses']) 
#         print('len num_g = ',num_g)
        if num_g>0:
            dist_g = np.round(response['businesses'][0]['distance'],0)
        else:
            dist_g = np.nan
        return num_g,dist_g


    else:
        ('all null',zipc,lat)
        num_g = np.nan
        dist_g=np.nan
        return num_g , dist_g



    
    return num_g,dist_g
```
