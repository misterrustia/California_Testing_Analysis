## join district level, housing and yelp data to scores and school level data

```python
cali= pd.read_csv('../../../data/processed/housing_cali_raw.csv')
```

```python
## load 2021 average housing price per zip

yr_avg=pd.read_csv('../../../data/processed/avg_housing_price_by_zip-00-21.csv')
yr_avg.head()
```

```python
## connect aveages to zipcodes in cali

cali = pd.concat([yr_avg['Avg_21'],cali],axis=1)
cali.iloc[:,:]
```

```python
## join housing to scores
```

```python
j1= dfs.merge(cali[['Zip_Code','Avg_21']],on='Zip_Code',how='left',indicator=True)
j1.drop('Unnamed: 0',inplace=True,axis=1)
print(j1['_merge'].value_counts())
j1.drop('_merge',inplace=True,axis=1)
```

## impute nulls from join with district mean than county mean 

```python
j2.drop(columns=['NSLP_Provision_Status'],inplace=True)
```

## impute unmatched district salar and pov numbers by county averages 

```python
cols = ['Total Salary Schedule FTE', 'Lowest Salary Offered',
       'Average Salary Paid', 'Highest Entry Level Step',
       'Highest Salary Offered', 'Salary at BA+60 Step 10',
      '2020–21 P‑2 ADA',
       '% Salary Change Over 2019–20','EDP 365',
       'Current\nExpense ADA', 'Current\nExpense Per ADA','Estimated Total Population', 'Estimated Population 5-17',
       'est_num_u17_pov']
```

```python

for i in cols:
    j2[i] = j2[i].fillna(j2.groupby(['County Code'])[i].transform('mean'))
j2[cols].info()
```

```python
j2.to_csv('../../../data/by_student_group/dfs_dist_housing.csv')

```
