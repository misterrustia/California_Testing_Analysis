
## Function to call and return data from yelp api

```python
 
def grocery_search(term,zipc,radius,lat,long):
    num_g = 0
    dist_g=0
    limit =50 
    if zipc >0:
        location = int(zipc)
    else:
        location = zipc 
        
    if lat  >0:
        try:
            response = yelp_api.search_query(term = term,
                                       latitude=lat,
                                     longitude=long,
                                     radius = radius,
                                     limit = search_limit)
        except:
            num_g = np.nan
            dist_g=np.nan
            return num_g , dist_g
            
        num_g = len(response['businesses'])
        if num_g>0:
            dist_g = np.round(response['businesses'][0]['distance'],0)
        else:
            dist_g = np.nan
        return num_g,dist_g
    
    elif location >0: # use zipcode 
        try:
            response = yelp_api.search_query(term = term,
                                             location=zipc,
                                     radius = radius,
                                     limit = search_limit)
        except:
            num_g = np.nan
            dist_g=np.nan
            return num_g , dist_g
        num_g = len(response['businesses']) 
        if num_g>0:
            dist_g = np.round(response['businesses'][0]['distance'],0)
        else:
            dist_g = np.nan
        return num_g,dist_g
    
    # fill in blanks will nan 
    else:
        num_g = np.nan
        dist_g=np.nan
        return num_g , dist_g


    return num_g,dist_g
```

```python

groceryt=[]
gt = []
term = 'grocery store'
search_limit = 50
radius = 5000 # in meters about 3 miles 
for i in range(len(test)):
    lat = test.iloc[i]['Latitude']
    long = test.iloc[i]['Longitude']
    zipc = test.iloc[i]['Zip Code']
   
    num_g,dist_g = grocery_search(term,zipc,radius,lat,long)
    
    groceryt.append(num_g)
    gt.append(dist_g)  
```
