## remove (#) and ed. and elem. and replace with matching from scores 

```python
names = s['LEA'].astype(str).values
final=[]
for name in names:
    if '(' in name:
        name = name.rsplit(' ',1)
        name = name[0]
#         print('final =',name)

        if  'ed.' in name:
            
            print(name)
            name = name.rsplit(' ',1)
            name = name[0]+' education'
#             print('final= ',name)
            final.append(name)
        elif 'elem.' in name:
            name = name.rsplit(' ',1)
            name = name[0]+' elementary'
#             print('final= ',name)
            final.append(name)
        else:
            final.append(name)
 
    elif  'ed.' in name:
#         print(name)
        name = name.rsplit(' ',1)
        name = name[0]+' education'
#         print('final= ',name)
        final.append(name)

    elif 'elem.' in name:
        name = name.rsplit(' ',1)
        name = name[0]+' elementary'
#         print('final= ',name)
        final.append(name)
            
    else:
        final.append(name)
        
s['LEA'] = final
```

## use mean of county number to fill blank estimate pop and pov numbers 

```python
cols =['Estimated Total Population', 'Estimated Population 5-17',
       'est_num_u17_pov','EDP 365', 'Current\nExpense ADA',
       'Current\nExpense Per ADA']

for i in cols:
    j4[i] = j4[i].fillna(j4.groupby(['Co'])[i].transform('mean'))
j4.info()
```
